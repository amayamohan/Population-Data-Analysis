```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/refs/heads/main/Task%201/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv'
#Skip the first 4 rows (which usually contain metadata or titles)
df = pd.read_csv(url , skiprows=4)
df.head()
```

```python
print("Data Types", df.dtypes)
```

```python
print("Shape of the data set", df.shape)
```

```python
print("Column Names", df.columns.tolist)
```

```python
print("Missing values per column ", df.isnull().sum())
```

✅ Step 1: Cleaning the Data
We first cleaned the dataset by:

Dropping unnecessary columns like 'Country Code'.

Removing rows with missing or non-numeric population values.

Converting the 2020 population column to numeric using pd.to_numeric.




```python
df_cleaned = df.drop(['Indicator Name', 'Indicator Code','Unnamed: 69'], axis=1)
df_cleaned.head()
```

```python
# Keep only rows with valid 2020 population
df_cleaned = df_cleaned.dropna(subset=['2020'])   #subset measn we only check a particular column
df_cleaned.head()
```

```python
df_cleaned[['Country Name', '2020']].isnull().sum()   #This means no missing values — perfect for plotting!
```

✅ Step 2: Histogram – Population Distribution
We plotted a histogram to show how many countries fall into specific population ranges.

```python
import matplotlib.pyplot as plt

# Convert population values to numeric (they might still be strings)
df_cleaned['2020'] = pd.to_numeric(df_cleaned['2020'], errors='coerce')

# Create the histogram
plt.figure(figsize=(10,6))
plt.hist(df_cleaned['2020'], bins=[0, 1e6, 10e6, 50e6, 100e6, 500e6, 1500e6], edgecolor='black')
plt.title('Distribution of Country Populations in 2020')
plt.xlabel('Population Range')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.xticks([0, 1e6, 10e6, 50e6, 100e6, 500e6, 1500e6],
           ['<1M', '1–10M', '10–50M', '50–100M', '100–500M', '500M–1.5B', '>1.5B'])


plt.show()
```

✅ Step 3: Bar Chart – Top 10 Most Populous Countries
We visualized the top 10 countries with the highest populations in 2020 using a bar chart.

```python
# Step 1: Sort the dataframe by population in 2020 in descending order
top_10 = df_cleaned.sort_values(by='2020', ascending=False).head(10)

# Step 2: Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10['Country Name'], top_10['2020'], color='orange', edgecolor='black')

# Step 3: Add titles and labels
plt.title('Top 10 Most Populous Countries in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Step 4: Show the chart
plt.show()
```
